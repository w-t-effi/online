```python
import pandas as pd
import os
```

```python
df= pd.DataFrame()
```

```python
for subdirs in os.walk('plots/spambase'):
    for subdir in subdirs[1]:
        df_=pd.read_csv('plots/spambase/'+subdir+'/accuracy.csv')
        df[subdir]=df_['Accuracy_scores']
        df_
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(df['n-F-x-0'],"-b", label="normalized/ x_drift_detection/ gamma=0")
plt.plot(df['n-F-x-0.2'],"-r", label="normalized/ x_drift_detection/ gamma=0.2")
plt.plot(df['n-F-y-0'],c='green', label="normalized/ y_drift_detection/ gamma=0")
plt.plot(df['n-F-y-0.2'],"-y", label="normalized/ y_drift_detection/ gamma=0.2")
plt.plot(df['u-F-y-0'],c='orange', label="unnormalized")
plt.plot(df['n_1s-F-y-0'],c='purple', label="normalized at the beginning")
#plt.ylim(0., 1.0)
plt.legend(loc="lower center")
plt.xlabel('Time Step', size=15, labelpad=1.6)
plt.ylabel('Accuracy', size=15, labelpad=1.5)
plt.title(f'Prediction Accuracy FIRES', size=20)
plt.savefig('prediction_accuracy_spambase_FIRES.png')#, bbox_inches='tight')
plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(df['n-S-x-0'],"-b", label="normalized/ x_drift_detection/ gamma=0")
plt.plot(df['n-S-x-0.2'],"-r", label="normalized/ x_drift_detection/ gamma=0.2")
plt.plot(df['n-S-y-0'],c='green', label="normalized/ y_drift_detection/ gamma=0")
plt.plot(df['n-S-y-0.2'],"-y", label="normalized/ y_drift_detection/ gamma=0.2")
plt.plot(df['u-S-y-0'],c='orange', label="unnormalized")
plt.plot(df['n_1s-S-y-0'],c='purple', label="normalized at the beginning")
plt.legend(loc="lower center")
plt.ylim(0.65, 1.0)
plt.xlabel('Time Step', size=15, labelpad=1.6)
plt.ylabel('Accuracy', size=15, labelpad=1.5)
plt.title(f'Prediction Accuracy SHAP', size=20)
plt.savefig('prediction_accuracy_spambase_SHAP.png')#, bbox_inches='tight')
plt.show()
plt.close()
```
